```python
import datetime

def to_timestamp(s):
    if s.split()[0].split('-')[0] == '18':
        return datetime.datetime.strptime(s, '%d-%m-%Y %H:%M')
    elif len(s.split()[1].split('.')) == 1:
        return datetime.datetime.strptime(s, '%Y-%m-%d %H:%M:%S')
    else:
        return datetime.datetime.strptime(s, '%Y-%m-%d %H:%M:%S.%f')
```

```python
cz['TimeStamp'] = cz['TimeStamp'].apply(to_timestamp)
cz
```

```python
import xml.etree.ElementTree as ET
from collections import Counter

class NoOSStringError(Exception):
    pass

def get_device(event):
    root = ET.fromstring(event)
    try:
        ua = user_agents.parse(root.attrib['OS'])
    except Exception as e:
        raise NoOSStringError()
    family = '' # ua.os.family
    if ua.is_tablet:
        family += 'Tablet'
    elif ua.is_mobile and not family.endswith('Phone'):
        family += 'Phone'
    elif ua.is_pc:
        family += 'PC'
    return family
    
```

```python
sessions['device'] = sessions['Content'].apply(get_device)
```

```python
respid_to_device = {}
for i, r in sessions[['KeyValue', 'device']].iterrows():
    respid_to_device[r['KeyValue']] = r['device']
```
